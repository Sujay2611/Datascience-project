```python
#Data Cleaning for IDS Project
#a)cleaning of %change column by replacing the _(underscore) with -ve sign only in those columns in which %change was -ve.
#b)Cleaning of Land_area column by deleting sq miles for each entity as we are interested in values relative to other values 
#and not in the absolute values.
#c)cleaning of Census and RecentCensus column by converting strings into integers in both columns.
#d)convertind %change from string into integer by removal of % symbol.

%matplotlib inline      
import matplotlib.pyplot as plt
import pandas as pd
from pandas.tools.plotting import scatter_matrix
import numpy as np
import seaborn as sns
import statistics as st
import re

df = pd.read_csv('city_stat_1.csv')


def isfloat(value):
    try:
        float(value)
        return True
    except:
        return False

for index,row in df.Percentage_change.iteritems():
    if(isfloat(row)):
        index=index+1
    else:
        if '_' in row:
            df.Percentage_change.loc[index]=row.split('_')
            df.Percentage_change.loc[index]="-"+df.Percentage_change.loc[index][1]
            print(df.Percentage_change.loc[index])
        else:
            print(row)
            

for index,row in df.Land_area.iteritems():
    df.Land_area.loc[index]=row.split(' ')
    df.Land_area.loc[index]=df.Land_area.loc[index][0]
    print(df.Land_area.loc[index])            
            
for index,row in df.RecentCensus.iteritems():
    df.RecentCensus.loc[index]=row.split(',')
    if len (df.RecentCensus.loc[index])==3:
        df.RecentCensus.loc[index]=df.RecentCensus.loc[index][0]+df.RecentCensus.loc[index][1]+df.RecentCensus.loc[index][2]
    elif len (df.RecentCensus.loc[index])==2:
        df.RecentCensus.loc[index]=df.RecentCensus.loc[index][0]+df.RecentCensus.loc[index][1]
    print( df.RecentCensus.loc[index])


for index,row in df.Census.iteritems():
    df.Census.loc[index]=row.split(',')
    if len (df.Census.loc[index])==3:
        df.Census.loc[index]=df.Census.loc[index][0]+df.Census.loc[index][1]+df.Census.loc[index][2]
    elif len (df.Census.loc[index])==2:
        df.Census.loc[index]=df.Census.loc[index][0]+df.Census.loc[index][1]
    print( df.Census.loc[index])
    
for index,row in df.Percentage_change.iteritems():
    df.Percentage_change.loc[index]=row.split('%')
    df.Percentage_change.loc[index]=df.Percentage_change.loc[index][0]
    

    
print(df)
```
